```python
def two_sum_1(nums, target):
    for i in range(len(nums)):
        for j in range(i + 1, len(nums)):
            if nums[i] + nums[j] == target:
                return [i, j]
```

```python
if __name__ == "__main__":
    print(two_sum_1([2, 7, 11, 15], 9))
    # expected: [0, 1]
```

```python
def two_sum_2(nums, target):
    for i,n in enumerate(nums):
        complement = target - n

        if complement in nums[i+1:]:
            return [nums.index(n), nums[i+1:].index(complement)+(i+1)]
        
print(two_sum_2([2,7,11,15], 9))
```

```python
def two_sum_3(nums, target):
    nums_map = {}
    for i, num in enumerate(nums):
        nums_map[num] = i

    for i, num in enumerate(nums):
        if target - num in nums_map and i != nums_map[target - num]:
            return [i, nums_map[target - num]]
        
print(two_sum_3([2, 7, 11, 15], 9))
```

```python
def two_sum_4(nums, target):
    nums_map = {}
    # combine in one for loop
    for i, num in enumerate(nums):
        if target - num in nums_map:
            return [nums_map[target - num], i]
        nums_map[num] = i

print(two_sum_4([2, 7, 11, 15], 9))
```
